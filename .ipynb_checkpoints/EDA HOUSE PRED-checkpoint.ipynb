# Feature Engineering Starts

```python
cat_feat_nan=[i for i in df.columns if df[i].isnull().sum()>0 and df[i].dtype=='O']
for i in cat_feat_nan:
    print(i,np.round(df[i].isnull().sum()),'% Missing Values')
```

Handling Categorical nan values

filling the missing value

```python
def fill_nan(df,cat_feat_nan):
    data=df.copy()
    data[cat_feat_nan]=data[cat_feat_nan].fillna('Missing')
    return data
df=fill_nan(df,cat_feat_nan)
df[cat_feat_nan].isnull().sum()
```

```python
df.head()
```

```python
#NUmerical Nan value
num_feat_nan=[i for i in df.columns if df[i].isnull().sum()>0 and df[i].dtype!='O']
for i in num_feat_nan:
    print(i,np.round(df[i].isnull().mean(),3),'% Missing values')
```

```python
#fill nan values 
for i in num_feat_nan:
    median_val=df[i].median()
    df[i+'NAN']=np.where(df[i].isnull(),1,0)
    df[i].fillna(median_val,inplace=True)
df[num_feat_nan].isnull().sum()
```

```python
df.head()
```

```python
df[year_feat].head()
```

```python
for i in ['YearBuilt','YearRemodAdd','GarageYrBlt']:
    df[i]=df['YrSold']-df[i]
df[year_feat].head()
```

```python
skewed_log_feat=['LotFrontage', 'LotArea', '1stFlrSF', 'GrLivArea', 'SalePrice']
for i in skewed_log_feat:
    df[i]=np.log1p(df[i])
```

```python
# Handling the rare feature less than 1%
cat_feat
```

```python
for i in cat_feat:
    data=df.groupby(i)['SalePrice'].count()/len(df)
    data_less_than_one=data[data<0.01].index
    df[i]=np.where(df[i].isin(data_less_than_one),'Rare variable',df[i])
```

```python
df.head(20)
```

# Handling Categorical features
This converts data to numerical data

```python
for i in cat_feat:
    labels_ordered=df.groupby(i)['SalePrice'].mean().sort_values().index
    labels_ordered={k:i for i,k in enumerate(labels_ordered,0)}
    df[i]=df[i].map(labels_ordered)
```

```python
df.head()
```

```python
# SCaling the data
scale_feat=[i for i in df.columns if i not in ['Id','SalePrice']]
print(len(scale_feat))
scale_feat
```

```python
from sklearn.preprocessing import MinMaxScaler
scalar=MinMaxScaler()
scalar.fit(df[scale_feat])
```

```python
scalar.transform(df[scale_feat])
```

```python
df1=pd.concat([df[["Id","SalePrice"]],pd.DataFrame(scalar.transform(df[scale_feat]),columns=scale_feat)],axis=1)
df1.head()
```

# feature Selection 

```python
from sklearn.linear_model import Lasso
from sklearn.feature_selection import SelectFromModel
```

```python
y=df1['SalePrice']
x=df1.drop(['Id','SalePrice'],axis=1)
```

```python
feat_sel=SelectFromModel(Lasso(alpha=0.005,random_state=0))
feat_sel.fit(x,y)
```

```python
feat_sel.get_support()
```

```python
selected_feat=x.columns[(feat_sel.get_support())]
print(len(selected_feat))
selected_feat
```

```python
x=x[selected_feat]
x.head()
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.01,random_state=0)
```

```python
from sklearn.tree import DecisionTreeRegressor
model=DecisionTreeRegressor(random_state=0)
model.fit(x_train,y_train)
```

```python
prediction=model.predict(x_test)
print("Train score",model.score(x_train,y_train))
print("test score",model.score(x_test,y_test))
```

```python
model1=Lasso(alpha=0.001,random_state=0)
model1.fit(x_train,y_train)
prediction1=model1.predict(x_test)
print("Train score",model1.score(x_train,y_train))
print("test score",model1.score(x_test,y_test))
```
