```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns
%matplotlib inline
```

```python
from sklearn.datasets import load_breast_cancer
cancer=load_breast_cancer()
```

```python
cancer.keys()
```

```python
df=pd.DataFrame(cancer.data,columns=cancer.feature_names)
df.head()
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
scaled_data=scaler.fit_transform(df)
```

```python
scaled_data
```

```python
da=pd.DataFrame(scaled_data,columns=cancer.feature_names)
da
```

```python
from sklearn.decomposition import PCA
pca=PCA(n_components=2,)
pca_da=pca.fit_transform(da)
pca_da
```

```python
plt.plot(pca.explained_variance_ratio_)
```

```python
plt.plot(np.cumsum(pca.explained_variance_ratio_))
```

```python
df.shape
```

```python
pca_da[0]
```

```python
pca_da.shape
```

```python
plt.figure(figsize=(8,6))
plt.scatter(pca_da[:,0],pca_da[:,1],c=cancer.target)
```

```python
plt.figure(figsize=(8,6))
sns.scatterplot(x=pca_da[:,0],y=pca_da[:,1],hue=cancer.target)
```

```python
plt.figure(figsize=(8,6))
plt.scatter(pca.components_[0],pca.components_[1])
```

```python
sns.heatmap(pca.components_)
```

```python
pca.components_.T.shape
```

```python
x_recover=pca.inverse_transform(pca_da)
```

```python
pd.DataFrame(x_recover,columns=cancer.feature_names)
```

```python
from sklearn.manifold import TSNE
tsne=TSNE()
tsne_d=tsne.fit_transform(scaled_data)

```

```python
tsne_d
```

```python
plt.figure(figsize=(20,20))
ax1=plt.subplot(2,2,1)
ax1.scatter(tsne_d[:,0],tsne_d[:,1],c=cancer.target)
ax2=plt.subplot(2,2,2)
ax2.scatter(pca_da[:,0],pca_da[:,1],c=cancer.target)
```
