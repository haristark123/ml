```python
import seaborn as sns
iris=sns.load_dataset('iris')
```

```python
iris.head()
```

```python
X=iris.drop(['species'],axis=1)
y=iris['species']
```

```python
from sklearn.decomposition import PCA
pca=PCA(n_components=2)
pca=pca.fit_transform(X)
```

```python
import pandas as pd
data=pd.DataFrame(pca,columns=['1_p','2_p'])
data.head()
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(10,5))
sns.scatterplot(pca[:,0],pca[:,1],hue=iris['species'],palette='Set1')
```

```python
from sklearn.mixture import GaussianMixture
model=GaussianMixture(n_components=3)
pred=model.fit_predict(data)
pred
```

```python
import matplotlib.pyplot as plt
iris['PCA1'] = pca[:, 0]
iris['PCA2'] = pca[:, 1]
iris['cluster']=pred
plt.figure(figsize=(10,5))
sns.lmplot('PCA1','PCA2',hue='species',col='cluster',data=iris,fit_reg=False)
```

```python
from sklearn.datasets import load_digits
digits=load_digits()
x=digits.data
y=digits.target
```

```python
x.shape
```

```python
from sklearn.manifold import Isomap
iso=Isomap(n_components=2)
data=iso.fit_transform(x)
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(10,10))
sns.scatterplot(data[:,0],data[:,1],hue=y,palette='Set1')
```

```python
from sklearn.decomposition import PCA
pca=PCA(n_components=2)
pca=pca.fit_transform(x)
plt.figure(figsize=(10,10))
plt.subplot(2,2,1)
sns.scatterplot(pca[:,0],pca[:,1],hue=y,palette='Set1')
plt.subplot(2,2,2)
sns.scatterplot(data[:,0],data[:,1],hue=y,palette='Set1')
```
